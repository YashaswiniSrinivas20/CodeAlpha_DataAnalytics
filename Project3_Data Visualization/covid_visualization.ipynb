```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("covid_data.csv")

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Line chart: Cases over time by country
sns.lineplot(data=df, x='date', y='cases', hue='country', marker="o")
plt.title("COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.show()

# Bar chart: Top 5 countries by cases on last date
latest_date = df['date'].max()
top_countries = df[df['date'] == latest_date].sort_values(by="cases", ascending=False)
sns.barplot(x="cases", y="country", data=top_countries, palette="viridis")
plt.title("Top Countries by COVID-19 Cases")
plt.show()

# Pie chart: Total Cases vs Deaths vs Recoveries
totals = df[["cases","deaths","recoveries"]].sum()
plt.pie(totals, labels=totals.index, autopct="%1.1f%%", startangle=90)
plt.title("Overall COVID-19 Distribution")
plt.show()
```
